```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data, Make sure to change to your working directory
data = pd.read_csv('D:\\01_Achieving a Neutral Carbon Footprint for Electrical Substations\\Codes Needing Upload\\SF6 Leak Plot of Breaker Pressure data_with_temperature_compensation.csv')

# Convert Timestamp column to datetime
data['Timestamp'] = pd.to_datetime(data['Timestamp'])

# Plot Uncompensated Breaker Pressure and Temperature Compensated Breaker Pressure Pressure over Time
plt.figure(figsize=(20, 6))
plt.plot(data['Timestamp'], data['Temperature Compensated SF6 Breaker Pressure (psig)'], label='Temperature Compensated SF6 Breaker Pressure', color='#00a0e1', marker='o')
plt.plot(data['Timestamp'], data['Uncompensated Pressure (psig)'], label='Uncompensated SF6 Breaker Pressure',color='#d7642c', marker='x', linestyle='--')
plt.xlabel('Time Horizon for the Observation')
plt.ylabel('SF6 Breaker Pressure (psig)')
plt.title('Uncompensated SF6 Breaker Pressure and Temperature Compensated SF6 Breaker Pressure Over Time')
plt.axhline(y=78, color='red', linestyle='--', label='Alarm Pressure (78 psig)')
plt.grid(True)
plt.legend()
plt.show()

# Plot Ambient Temperature over Time
#plt.figure(figsize=(20, 6))
#plt.plot(data['Timestamp'], data['Ambient Temperature (F)'], label='Ambient Temperature', color='orange', marker='o')
#plt.xlabel('Time')
#plt.ylabel('Temperature (F)')
#plt.title('Ambient Temperature Fluctuation Over Time')
#plt.grid(True)
#plt.legend()
#plt.show()

# Plot both Uncompensated Breaker Pressure and Temperature Compensated Breaker Pressure on the same plot
fig, ax1 = plt.subplots(figsize=(16, 8))
ax1.set_xlabel('Time Horizon for the Observation')
ax1.set_ylabel('SF6 Breaker Pressure (psig)', color='black')
ax1.plot(data['Timestamp'], data['Temperature Compensated SF6 Breaker Pressure (psig)'], label='Temperature Compensated Breaker Pressure', color='#00a0e1', marker='o')
ax1.plot(data['Timestamp'], data['Uncompensated Pressure (psig)'], label='Uncompensated Pressure (psig)', color='#d7642c', marker='x', linestyle='--')
ax1.axhline(y=78, color='red', linestyle='--', label='Alarm Pressure (78 psig)')
ax2 = ax1.twinx()
ax2.set_ylabel('Ambient Temperature (F)', color='black')
ax2.plot(data['Timestamp'], data['Ambient Temperature (F)'], label='Ambient Temperature (F)', color='lightsteelblue', marker='x')
fig.tight_layout()
plt.title('Uncompensated SF6 Breaker Pressure and Temperature Compensated SF6 Breaker Pressure Over Time')
ax1.legend(loc='lower left')
ax2.legend(loc='upper right')
plt.savefig('D:\\01_Achieving a Neutral Carbon Footprint for Electrical Substations\\Figures\\my_plot.png')
```
